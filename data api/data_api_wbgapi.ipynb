# Extracción de Datos de la API del Banco Mundial

```python
# importar librerías
import wbgapi as wb
import pandas as pd
```

```python
# crear listas de países e indicadores
country_list = [
                "AFG", "ARE", "ARG", "AUS", "AUT", "BFA", "BRA", "CAN", "CHE", "CHN",
                "COL", "CRI", "CUB", "DEU", "DNK", "DZA", "ECU", "EGY", "ESP", "FIN", 
                "FJI", "GBR", "GRC", "IND", "JPN", "KEN", "KHM", "MAR", "MEX", "NER", 
                "NLD", "NOR", "NPL", "NZL", "RUS", "RWA", "SAU", "SGP", "SLE", "TCD", 
                "USA", "VEN", "WSM", "ZAF"                
               ]

indicators_list = [
                 "SP.DYN.LE00.IN", "SP.DYN.LE00.FE.IN", "SP.DYN.LE00.MA.IN", "SP.POP.GROW",
                 "SP.RUR.TOTL.ZG", "SP.URB.GROW", "SP.POP.TOTL", "SP.URB.TOTL", "SP.RUR.TOTL"
                 "SP.POP.TOTL.MA.IN", "SP.POP.TOTL.FE.IN", "SP.POP.0014.TO", "SP.POP.0014.MA.IN"
                 "SP.POP.0014.FE.IN", "SP.POP.1564.TO", "SP.POP.1564.MA.IN", "SP.POP.1564.FE.IN",
                 "SP.POP.65UP.TO", "SP.POP.65UP.MA.IN", "SP.POP.65UP.FE.IN", "NY.GDP.PCAP.KD.ZG"
                 "NY.GDP.PCAP.CD", "NY.GDP.DEFL.KD.ZG", "FP.CPI.TOTL.ZG", "SE.SEC.CUAT.LO.ZS",
                 "SE.SEC.CUAT.UP.ZS", "SE.SEC.CUAT.PO.ZS", "SE.SEC.CMPT.LO.ZS", "SH.XPD.GHED.PC.CD"
                 "SH.XPD.OOPC.PC.CD", "SH.XPD.PVTD.PC.CD"
                  ]
```

```python
# Obtener los datos de la API del Banco Mundial
df_api = wb.data.DataFrame(indicators_list, country_list, mrv=36).reset_index()
```

```python
df_api
```

```python
# obtener los nombres de los países e indicadores
countries_info = wb.economy.info().table()
indicators_info = wb.series.info().table()
```

```python
# Mapear los nombres de los países y los nombres de los indicadores en el DataFrame
df_api['country_name'] = df_api['economy'].map(lambda x: next((country[1] for country in countries_info if country[0] == x), ''))
df_api['indicator_name'] = df_api['series'].map(lambda x: next((indicator[1] for indicator in indicators_info if indicator[0] == x), ''))
```
