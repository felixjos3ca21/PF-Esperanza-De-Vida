```python
# importar librerías
import pandas as pd
import wbdata as wb
```

```python
# Definir listas de indicadores y países
indicators_list = [
                     "SP.DYN.LE00.IN", "SP.DYN.LE00.FE.IN", "SP.DYN.LE00.MA.IN", "SP.POP.GROW",
                     "SP.RUR.TOTL.ZG", "SP.URB.GROW", "SP.POP.TOTL", "SP.URB.TOTL", "SP.RUR.TOTL"
                     "SP.POP.TOTL.MA.IN", "SP.POP.TOTL.FE.IN", "SP.POP.0014.TO", "SP.POP.0014.MA.IN"
                     "SP.POP.0014.FE.IN", "SP.POP.1564.TO", "SP.POP.1564.MA.IN", "SP.POP.1564.FE.IN",
                     "SP.POP.65UP.TO", "SP.POP.65UP.MA.IN", "SP.POP.65UP.FE.IN", "NY.GDP.PCAP.KD.ZG"
                     "NY.GDP.PCAP.CD", "NY.GDP.DEFL.KD.ZG", "FP.CPI.TOTL.ZG", "SE.SEC.CUAT.LO.ZS",
                     "SE.SEC.CUAT.UP.ZS", "SE.SEC.CUAT.PO.ZS", "SE.SEC.CMPT.LO.ZS", "SH.XPD.GHED.PC.CD"
                     "SH.XPD.OOPC.PC.CD", "SH.XPD.PVTD.PC.CD"
                  ]
country_list = [
                    "AFG", "ARE", "ARG", "AUS", "AUT", "BFA", "BRA", "CAN", "CHE", "CHN",
                    "COL", "CRI", "CUB", "DEU", "DNK", "DZA", "ECU", "EGY", "ESP", "FIN", 
                    "FJI", "GBR", "GRC", "IND", "JPN", "KEN", "KHM", "MAR", "MEX", "NER", 
                    "NLD", "NOR", "NPL", "NZL", "RUS", "RWA", "SAU", "SGP", "SLE", "TCD", 
                    "USA", "VEN", "WSM", "ZAF"                
               ]
```

```python
# Obtener nombres de indicadores
indicators_names = {}
all_indicators = wb.get_indicator(source=2)
for indicator in indicators_list:
    matching_indicators = [i for i in all_indicators if i['id'] == indicator]
    if matching_indicators:
        indicators_names[indicator] = matching_indicators[0]['name']

# Obtener nombres de países
countries_names = {}
for country in country_list:
    info = wb.get_country(country_id=country)
    countries_names[country] = info[0]['name']
```

```python
# visualizar los diccionarios creados
print(indicators_names)
countries_names
```

```python
# crear dataframe con los diccionarios creados y con frecuencia anual
data = wb.get_dataframe(indicators=indicators_names, country=countries_names, freq='Y')
```

```python
# visualizar el dataframe creado
data
```

```python
# Restablecer el índice del DataFrame
df_api = data.reset_index()
```

```python
# visualizar el dataframe resultante
df_api
```

```python
# eliminar los años anteriores a los años escogidos
df_api_final = df_api[~df_api['date'].astype(str).str[:4].between('1960', '1986')]
```

```python
# resetear el índice
df_api_final = df_api_final.reset_index(drop=True)
```
